```python
# Create categories for the expected datatype in each category
numeric_cols = ['Bottle_Volume_ml', 'State_Bottle_Cost', 'State_Bottle_Retail', 'Bottles_Sold', 'Sale_Dollars', 'Volume_Sold_Liters', 'Volume_Sold_Gallons']
cat_cols = ['Invoice_Item_Number', 'Store_Number', 'Store_Name', 'Address', 'City', 'Zip_Code', 'Store_Location', 'County_Number', 'County', 'Category', 'Category_Name', 'Vendor_Number', 'Vendor_Name', 'Item_Number', 'Item_Description', 'Pack']
date_cols = ['Date']
```

```python
# Create chunks to process the dataset 100,000 rows at a time
chunksize = 100000
chunks = pd.read_csv('cleaned_2019_2025.csv', chunksize=chunksize, dtype_backend='numpy_nullable')
cleaned_chunks = []
```

```python
# Handle missing values for each category
for chunk in chunks:
    # Numeric: Impute with median
    for col in numeric_cols:
        median_val = chunk[col].median()
        chunk[col] = chunk[col].fillna(median_val)
    
    # Categorical: Replace NaN/empty with 'UNKNOWN'
    for col in cat_cols:
        chunk[col] = chunk[col].astype(str).replace('nan', 'UNKNOWN').replace('', 'UNKNOWN').str.strip()
    
    # Date: Parse and impute
    chunk['Date'] = pd.to_datetime(chunk['Date'], errors='coerce', format='%m/%d/%Y')
    min_date = chunk['Date'].min()
    chunk['Date'] = chunk['Date'].fillna(min_date)
    
    cleaned_chunks.append(chunk)
```

```python
# Combine the chunks that were processed
df = pd.concat(cleaned_chunks, ignore_index=True)
```

```python
# Handle infinite values
for col in numeric_cols:
    df[col] = df[col].replace([np.inf, -np.inf], np.nan)
    df[col] = df[col].fillna(df[col].median())
```

```python
# Normalize date format
df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%Y-%m-%d')
```

```python
# Define chunk size for large dataset
chunksize = 100000

# Initialize chunk processing
chunks = pd.read_csv('cleaned_2019_2025.csv', chunksize=chunksize, dtype_backend='numpy_nullable')
cleaned_chunks = []

# Process each chunk
for chunk in chunks:
    # Step 1: Convert 'County Number' to string to handle mixed types
    if 'County_Number' in chunk.columns:
        chunk['County_Number'] = chunk['County_Number'].astype(str)
        
        # Step 2: Replace 'Na', 'NaN', or 'nan' with '0'
        chunk['County_Number'] = chunk['County_Number'].replace(['Na', 'NaN', 'nan'], '0')
        
        # Step 3: Remove '.0' from float-like strings (e.g., '77.0' -> '77')
        chunk['County_Number'] = chunk['County_Number'].str.replace(r'\.0$', '', regex=True)
        
        # Step 4: Convert to numeric, coerce errors to NaN, then fill with 0 and convert to int
        chunk['County_Number'] = pd.to_numeric(chunk['County_Number'], errors='coerce').fillna(0).astype(int)
    
    cleaned_chunks.append(chunk)

# Combine chunks
df = pd.concat(cleaned_chunks, ignore_index=True)
```

```python
import pandas as pd
import re
import logging
import gc
import sys

### This set of instructions to convert store location from 'Point (-92.811539 43.285134)' into longitude and latitude. ###
# Set up logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s [%(levelname)s] %(message)s',
    handlers=[
        logging.FileHandler('cleaning_log.txt'),
        logging.StreamHandler(sys.stdout)
    ]
)

def parse_point(location):
    """
    Parse 'Store Location' string (e.g., 'Point (-92.811539 43.285134)') into longitude and latitude.
    Returns (longitude, latitude) as floats or (None, None) for invalid/missing values.
    """
    if pd.isna(location) or location == "" or str(location).lower() in ["na", "nan"] or str(location).strip() == "":
        return None, None
    
    # Regex to match 'POINT (lon lat)' or '(lon, lat)' or '(lon lat)'
    pattern = r'(?:POINT\s*)?\(?([-]?\d+\.\d+)[, ]+([-]?\d+\.\d+)\)?'
    try:
        match = re.match(pattern, str(location).strip())
        if match:
            longitude = float(match.group(1))
            latitude = float(match.group(2))
            return longitude, latitude
        return None, None
    except Exception as e:
        logging.error(f"Error parsing location '{location}': {e}")
        return None, None

# Parameters
input_file = 'cleaned_2019_2025part2.csv'
output_file = 'final_cleaned_2019_2025.csv'
chunksize = 100000

# Initialize variables
first_chunk = True
chunk_count = 0
total_rows = 0

logging.info(f"Starting processing of {input_file}")

try:
    # Process dataset in chunks
    chunks = pd.read_csv(
        input_file,
        chunksize=chunksize,
        dtype_backend='numpy_nullable',
        encoding='utf-8',  # Explicit encoding to handle potential issues
        low_memory=False   # Prevent mixed type warnings
    )
    
    for chunk in chunks:
        chunk_count += 1
        logging.info(f"Processing chunk {chunk_count} with {len(chunk)} rows")
        
        try:
            # Apply parse_point to 'Store_Location'
            chunk[['Longitude', 'Latitude']] = chunk['Store_Location'].apply(
                lambda x: pd.Series(parse_point(x))
            )
            
            # Save chunk to output file
            mode = 'w' if first_chunk else 'a'
            header = first_chunk
            chunk.to_csv(
                output_file,
                mode=mode,
                header=header,
                index=False,
                encoding='utf-8'
            )
            
            total_rows += len(chunk)
            first_chunk = False
            logging.info(f"Saved chunk {chunk_count} to {output_file}")
            
            # Clear memory
            del chunk
            gc.collect()
            
        except Exception as e:
            logging.error(f"Error processing chunk {chunk_count}: {e}")
            raise
    
    logging.info(f"Completed processing {chunk_count} chunks with {total_rows} rows")
    logging.info(f"Output saved to {output_file}")

except Exception as e:
    logging.error(f"Fatal error: {e}")
    sys.exit(1)
```
